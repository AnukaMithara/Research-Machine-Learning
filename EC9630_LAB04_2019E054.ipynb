### 1. You are requested to use HCV dataset directly from UCI data repository. However, you can use any classification dataset for this purpose.

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn import preprocessing
from sklearn.metrics import accuracy_score

df = pd.read_csv('hcv_dataset.csv')

df
```

### 2. Prepare your dataset for learning.


```python

# Drop any rows with missing values
df = df.dropna(axis=0)
df
```

```python
# Convert all object columns to categorical data
for col in df.select_dtypes(include='object'):
    df.loc[:, col] = df[col].astype('category')
    df.loc[:, col] = df[col].cat.codes

df
```

```python

# Separate the features and target variable

X = df.drop("Category", axis=1)
y = df["Category"]

df
```

### 3. First, Split your dataset into training set and test set. Build SVM using following kernels and measure the performance of your model.
#### (a) Linear Kernel
#### (b) Polynomial Kernel
#### (c) RBF kernel

```python
# Split the dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=54)
```

```python
# Initialize the SVM classifiers with different kernels
linear_svm = SVC(kernel='linear')
poly_svm = SVC(kernel='poly')
rbf_svm = SVC(kernel='rbf')

# Fit the models on the training data
linear_svm.fit(X_train, y_train)
poly_svm.fit(X_train, y_train)
rbf_svm.fit(X_train, y_train)

# Predict the classes for test set
linear_predictions = linear_svm.predict(X_test)
poly_predictions = poly_svm.predict(X_test)
rbf_predictions = rbf_svm.predict(X_test)

# Measure the performance of the models
linear_accuracy = accuracy_score(y_test, linear_predictions)
poly_accuracy = accuracy_score(y_test, poly_predictions)
rbf_accuracy = accuracy_score(y_test, rbf_predictions)

# Print the accuracies
print("Linear Kernel Accuracy\t\t:", linear_accuracy)
print("Polynomial Kernel Accuracy\t:", poly_accuracy)
print("RBF Kernel Accuracy\t\t:", rbf_accuracy)
```

### 4. Now, do the same exercise using 10-fold and 5-fold cross validation and interpret your result in the proper way

```python
# Perform 10-fold cross-validation
linear_scores = cross_val_score(linear_svm, X_train, y_train, cv=10)
poly_scores = cross_val_score(poly_svm, X_train, y_train, cv=10)
rbf_scores = cross_val_score(rbf_svm, X_train, y_train, cv=10)

# Compute the mean accuracy and standard deviation
linear_mean_accuracy = linear_scores.mean()
poly_mean_accuracy = poly_scores.mean()
rbf_mean_accuracy = rbf_scores.mean()


# Print the mean accuracy and standard deviation for each kernel
print("------------------------------------------------------------------------")
print("10-fold cross validation")
print("Linear Kernel Mean Accuracy\t:", linear_mean_accuracy)
print("Polynomial Kernel Mean Accuracy\t:", poly_mean_accuracy)
print("RBF Kernel Mean Accuracy\t:", rbf_mean_accuracy)
print("------------------------------------------------------------------------")


# Perform 5-fold cross validation
linear_scores = cross_val_score(linear_svm, X_train, y_train, cv=5)
poly_scores = cross_val_score(poly_svm, X_train, y_train, cv=5)
rbf_scores = cross_val_score(rbf_svm, X_train, y_train, cv=5)

# Compute the mean accuracy and standard deviation
linear_mean_accuracy = linear_scores.mean()
poly_mean_accuracy = poly_scores.mean()
rbf_mean_accuracy = rbf_scores.mean()

# Print the mean accuracy and standard deviation for each kernel
print("------------------------------------------------------------------------")
print("5-fold cross validation")
print("Linear Kernel Mean Accuracy\t:", linear_mean_accuracy)
print("Polynomial Kernel Mean Accuracy\t:", poly_mean_accuracy)
print("RBF Kernel Mean Accuracy\t:", rbf_mean_accuracy)
print("------------------------------------------------------------------------")
```
